## Finding the average salary of different job roles

```python
data_bar = data[['JobRole','MonthlyIncome']]
data_bar = data_bar.groupby(['JobRole'], as_index=False).mean()
data_bar
```

## Barchart using matplotlib libraries


```python
data_bar.plot.barh(x='JobRole', y='MonthlyIncome')
plt.xlabel('Average Monthly Income') # add to x-label to the plot
plt.ylabel('Job Role') # add y-label to the plot
plt.title('Average Monthly Income Vs Job Role') # add title to the plot
plt.legend(['Average Monthly Income'])
plt.show()
```

## Barchart using seaborn

```python

#x = data[["JobRole"]]
#y = data[["MonthlyIncome"]]
sns.barplot(x='MonthlyIncome', y='JobRole', data=data_bar)

# Set axis labels and title
plt.xlabel('Monthly Income')
plt.ylabel('Job Role')
plt.title('Horizontal Bar Chart Example')
plt.legend(title='Average Monthly Income')
# Display the chart
plt.show()





```

## Bar Chart using Plotly

```python
import plotly.express as px

# Data for the chart
job_roles = data_bar['JobRole']
monthly_incomes = data_bar['MonthlyIncome']

# Create a DataFrame for the chart
data = {'Job Role': job_roles, 'Monthly Income': monthly_incomes}
df = pd.DataFrame(data)

# Create the horizontal bar chart
fig = px.bar(df, x='Monthly Income', y='Job Role', orientation='h', color='Monthly Income', title='Average Monthly Income Vs Job Role')

# Display the chart
fig.show()

```

## Summary
After comparing the 3 charts,  the barchart using seaborn is more visually appealing and easy to interpret data compared to plotly and matplotlib. Plotly is also a better choice compared to matplotlib.
The chart allows us to quickly see which job roles have higher or lower average incomes, and we can easily compare the incomes of different job roles.
